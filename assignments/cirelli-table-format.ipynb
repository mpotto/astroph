# Formatting *Cirelli* table 

```python
import numpy as np
import pandas as pd
```

```python
spectras_table = 'spectras/AtProduction_gammas.dat'
```

```python
columns= ["$m_{DM}$", "$\log_{10} x$", "$e^L$", "$e^R$", "$e$", "$\mu^L$", "$\mu^R$", "$\mu$", "$\tau^L$", "$\tau^R$", r"$\tau$", r"$q\bar{q}$", r"$c\bar{c}$", r"$b\bar{b}$",
          r"$t\bar{t}$", "$W^L$", "$W^T$", "$W$", "$Z^L$", "$Z^T$", "$Z$", r"$g\bar{g}$", "$\gamma$", r"$h$", "$\nu_e$", r"$\nu_{\mu}$", r"$\nu_{\tau}$", 
          r"$VV \rightarrow 4e$", r"$VV \rightarrow 4\mu$", r"$VV \rightarrow 4\tau$"]
```

```python
data = np.loadtxt(spectras_table, skiprows=1)
```

```python
spectras = pd.DataFrame(data, columns=columns)
```

```python
spectras.tail()
```

```python
spectras.to_pickle('spectras/at-production-gammas.pkl')
```
